$$
\begin{align}
    \textrm{conv}(S)  =  \{  \sum_{i = 1}^n \lambda_{i} x_i | x_i \in S, \lambda_{i} > 0, \sum_{i = 1}^n \lambda_{i} =1 \}
\end{align}
$$

```python
data = np.random.randn(100,2)
x =  data[:,0]
y = data[:,1]
```

```python
hull = ConvexHull(data)
```

```python
plt.plot(x,y,'o',color='red')
for simplex in hull.simplices:
    plt.plot(data[simplex,0],data[simplex,1],'b-')
plt.show()
```

```python
def plot_hull(n):
    data = np.random.randn(n,2)
    x =  data[:,0]
    y = data[:,1]
    hull = ConvexHull(data)
    plt.plot(x,y,'o',color='red')
    for simplex in hull.simplices:
        plt.plot(data[simplex,0],data[simplex,1],'b-')
    plt.show()
```

```python
plot_hull(1000)
```

```python
plot_hull(200)
```

```python
plot_hull(1000000)
```

```python
plot_hull(100)
```

```python
x = np.linspace(-5, 5, 1000)
y = np.linspace(-5, 5, 1000)
z = np.linspace(-5, 5, 1000)
```

```python
X, Y,Z  = np.meshgrid(x, y,z, sparse=True)
```

```python
f = lambda x,y,z: np.sin(x**2 + y**2) / np.exp(x**2 + z**2)
```

```python
out =  f(X,Y,Z)
```

```python
plt.imshow(out[-1])
```

```python
fig, ax = plt.subplots() # creating my fig
camera = Camera(fig)# the camera gets the fig we'll plot
for i in range(out.shape[2]):
    ax.imshow(out[i])
#     ax.plot([i]* 5, c='red',linewidth=2) # 5 element array from 0 to 9
    camera.snap() # the camera takes a snapshot of the plot
animation = camera.animate(blit=False, interval=40) # animation ready
HTML(animation.to_html5_video()) # displaying the animation
```

```python
animation.save('animation_function.mp4',dpi=300,savefig_kwargs={'frameon': False,'pad_inches': 'tight'})
```

```python
x = np.linspace(-1,1,1000)
y = np.linspace(-1,1,1000)

X,Y =  np.meshgrid(x,y)

a =  np.random.normal(0,1,(1000,))
b =  np.random.normal(0,1,(1000,))


fig = plt.figure(1,figsize=[12,8])

ax = fig.add_subplot(projection='3d')

Z =  (X**2/a[0]) + (Y**2/ b[0])
surf = ax.plot_surface(X,Y,Z)
```
