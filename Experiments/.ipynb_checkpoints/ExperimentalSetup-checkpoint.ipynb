```python
file_path =  'data/clip_loango.npz'
# file_path2 =  '/data/clip_loango_resized.npz'
FRAME_LENGHT =  1024
SAMPLE_RATE =  44100
HOP_LENGTH =  512
```

```python
# CLASSES =  get_classes(INPUT_PATH)
```

```python
data =  np.load(file_path, allow_pickle=True)
X = data['x']
y = data['y']
Z =  data['z']
CLASSES = list(data['c'])
```

```python
def display(X,n=6,flag =  False):
  idx =  np.random.randint(0, X.shape[0], n)
  plt.figure(1,figsize=(24,8))
  for i in range(len(idx)):
    plt.subplot(2,3,i+1)
    if(flag ==  False):
      librosa.display.waveplot(Z[idx[i]],sr=SAMPLE_RATE,alpha=0.6)
      plt.xlabel("time")
      plt.ylabel("amplitude")
    else:
      librosa.display.specshow(X[idx[i]],sr=SAMPLE_RATE,y_axis='mel', x_axis='time', hop_length=HOP_LENGTH//2)
    plt.title(CLASSES[y[idx[i]]])
  plt.tight_layout()
  if(flag ==  False):
    plt.savefig(f'{figures}/Sample_Waveplot.png', bbox_inches ="tight", dpi=300)
  else:
    plt.savefig(f'{figures}/Sample_PowerSpectrum.png', bbox_inches ="tight", dpi=300)
  plt.show()
```

```python
input_shape =  (X.shape[1],X.shape[2],1)
output =  len(CLASSES)
X = X[...,np.newaxis]
```

```python
X = (X -  X.mean(axis=0))/ X.std(axis=0)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.1, shuffle=True)
```

```python
pca = PCA(n_components=3)
X_pca =  pca.fit_transform(X.reshape(X.shape[0],-1))
```

```python
def plot_components(X_pca, x_str='Component 1', y_str='Component 2', str_title="PCA", title="title.pdf"):
  fig = plt.figure(1,figsize=(10,6))
  ax =  fig.add_subplot(111)
  scatter = ax.scatter(X_pca[:,0],X_pca[:,1], c=list(y))
  ax.set_xlabel(x_str)
  ax.set_ylabel(y_str)
  ax.set_title(str_title)
  # ax.set_zlabel("PCA Component 3")
  ax.legend(handles=scatter.legend_elements()[0], labels=CLASSES,bbox_to_anchor=(1.0, 1.0))
  plt.savefig(f'{figures}/{title}', bbox_inches ="tight", dpi=100)
  plt.show()
```

```python
plot_components(X_pca,x_str='PCA 1', y_str='PCA 2', str_title="Principal Component Analysis", title="Normalized_PCA.png")
```

```python
svd = TruncatedSVD(n_components=3)
X_svd =  svd.fit_transform(X.reshape(X.shape[0],-1))
```

```python
plot_components(X_svd,x_str='SVD 1', y_str='SVD 2', str_title="Singular Value Decomposition", title="Normalized_SVD.png")
```

```python
ica =  FastICA(n_components=3, random_state=np.random.seed(42))
X_ica = ica.fit_transform(X.reshape(X.shape[0],-1))
```

```python
plot_components(X_ica,x_str='ICA 1', y_str='ICA 2', str_title="Independent Component Analysis", title="Normalized_ICA.png")
```

```python
idx_a,idx_b = np.random.randint(0,Z.shape[0], 2)
D = librosa.amplitude_to_db(np.abs(librosa.stft(Z[idx_a],hop_length=HOP_LENGTH)), ref=np.max)
D2 = librosa.amplitude_to_db(np.abs(librosa.stft(Z[idx_b],hop_length=HOP_LENGTH)), ref=np.max)

plt.figure(figsize=(16,6))
plt.subplot(2,2,1)
plt.title(f"Original Signal for {CLASSES[y[idx_a]]}")
librosa.display.specshow(D, y_axis='log',sr=SAMPLE_RATE,hop_length=HOP_LENGTH, x_axis='time')

plt.subplot(2,2,2)
plt.title(f"PCEN Normalized Original Signal for {CLASSES[y[idx_a]]}")
librosa.display.specshow(librosa.pcen(np.abs(librosa.stft(Z[idx_a],hop_length=HOP_LENGTH)), sr=SAMPLE_RATE, hop_length=HOP_LENGTH), y_axis='log',x_axis='time',sr=SAMPLE_RATE,hop_length=HOP_LENGTH)
plt.subplot(2,2,3)
plt.title(f"Original Signal for {CLASSES[y[idx_b]]}")
librosa.display.specshow(D2, y_axis='log',sr=SAMPLE_RATE,hop_length=HOP_LENGTH, x_axis='time')
plt.subplot(2,2,4)
plt.title(f"PCEN Normalized Signal for {CLASSES[y[idx_b]]}")
librosa.display.specshow(librosa.pcen(np.abs(librosa.stft(Z[1],hop_length=HOP_LENGTH)), sr=SAMPLE_RATE, hop_length=HOP_LENGTH), y_axis='log',x_axis='time',sr=SAMPLE_RATE,hop_length=HOP_LENGTH)
plt.tight_layout()
plt.savefig(f'{figures}/spectral_normalization_PCEN.png', bbox_inches ="tight", dpi=100)
plt.show()
```

```python
# First import everthing you need
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D


n = len(X_pca[:,0])
xx = X_pca[:,0]
yy = X_pca[:,1]
zz = X_pca[:,2]

# Create a figure and a 3D Axes
fig = plt.figure(3, figsize=(10,6))
fig.patch.set_facecolor(None)
fig.patch.set_alpha(0.0)
ax = Axes3D(fig)
ax.grid(False)
ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False
ax.axis('off')
# Create an init function and the animate functions.
# Both are explained in the tutorial. Since we are changing
# the the elevation and azimuth and no objects are really
# changed on the plot we don't have to return anything from
# the init and animate function. (return value is explained
# in the tutorial.
def init():
    scatter = ax.scatter(xx, yy, zz, marker='o', s=20, c=list(y), alpha=0.6)
    ax.set_xlabel("PCA 1")
    ax.set_ylabel("PCA 2")
    ax.set_zlabel("PCA 3")
    ax.legend(handles=scatter.legend_elements()[0], labels=CLASSES,bbox_to_anchor=(1.2, 0.8))
    return fig,

def animate(i):
    ax.view_init(elev=10., azim=i)
    return fig,

# Animate
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=360, interval=20, blit=True)
```

```python
# Save
anim.save('./Figures/animation_PCA.mp4',dpi=300,savefig_kwargs={'frameon': False,'pad_inches': 'tight'})
```
