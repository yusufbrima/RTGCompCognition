```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import os
```

```python
# We are setting global constants

FRAME_LENGHT =  1024
HOP_LENGTH =  512
```

```python
def get_files(path):
    # pp = "/net/projects/scratch/winter/valid_until_31_July_2022/krumnack/animal-communication-data/Chimp_IvoryCoast/manually_verified_2s/chimp_only_23112020_with_ids"
    files = [os.path.join(path,f) for f in os.listdir(path) if os.path.isfile(os.path.join(path,f)) and f.endswith('.wav')]
    return files

def compute_fft(y):
    ft =  sp.fft.fft(y)
    magnitude =  abs(ft)
    freq =  np.linspace(0,sr,len(magnitude))
    return (magnitude,freq)

def compute_mfc(y,sr,n_mels=128):
    mel_spectrogram = librosa.feature.melspectrogram(y,sr=sr,n_fft=FRAME_LENGHT,hop_length=HOP_LENGTH,n_mels=n_mels)
    log_mel_spectrogram =  librosa.power_to_db(mel_spectrogram,ref=np.max)
    return log_mel_spectrogram


def amplitude_envelope_overlap(signal,frame_size=FRAME_LENGHT,hop_length=HOP_LENGTH):
    return np.array([max(signal[i:i+frame_size] )  for i in range(0,signal.size,hop_length) ])


def set_figure_paths(dirname):
    if(os.path.isdir(os.path.join(os.getcwd(),dirname))):
        figpath =  os.path.join(os.getcwd(),dirname)
    else:
        os.mkdir(os.path.join(os.getcwd(),dirname))
        figpath =  os.path.join(os.getcwd(),dirname)
    return figpath


def plot_graph(y,type=1,title="",save=False,figsize=(10,4)):
    plt.figure(1,figsize=figsize)
    if type ==1:
        # the code plots a waveplot for a signal y at sampling rate sr
        librosa.display.waveplot(y,alpha=0.5)
    elif type == 2:
        # the code segment computes the power spectrum of a digital signal
        m,f=compute_fft(y)
        plt.plot(f,m)
        plt.xlabel("Frequency (Hz)")
        plt.ylabel("Magnitude")
        plt.xticks(np.arange(min(f), max(f)+1, 1500.0))
        plt.xlim([0,max(f)])
    elif type == 3:
        # Power Spectrogram 
        D = librosa.amplitude_to_db(np.abs(librosa.stft(y,n_fft=FRAME_LENGHT,hop_length=HOP_LENGTH)), ref=np.max)
        img = librosa.display.specshow(D, y_axis='log',sr=sr,hop_length=HOP_LENGTH)
        plt.colorbar(format='%+2.0f dB')
    elif type == 4:
        #Constant-Q power spectrogram 
        CQT = librosa.amplitude_to_db(np.abs(librosa.cqt(y, sr=sr)), ref=np.max)
        librosa.display.specshow(CQT, y_axis='cqt_note')
        plt.colorbar(format='%+2.0f dB')
    elif type ==  5:
        # Mel Frequency Spectrogram
        log_mel_spectrogram = compute_mfc(y,sr,n_mels=128)
        librosa.display.specshow(log_mel_spectrogram, x_axis='time',y_axis="mel",sr=sr)
        plt.colorbar(format='%+2.f')
    elif type == 6:
        #Mel-frequency Cepstral Coefficient
        mfccs = librosa.feature.mfcc(y=y, sr=sr,n_fft=FRAME_LENGHT,hop_length=HOP_LENGTH,n_mfcc=64) #librosa.feature.mfcc(y=y, sr=sr,n_fft=FRAME_LENGHT,hop_length=HOP_LENGTH,n_mels=256)
        librosa.display.specshow(mfccs, sr=sr, x_axis='time',y_axis="log")
        plt.colorbar(format='%+2.f')
    elif type == 7:
        mfccs = librosa.feature.mfcc(y=y, sr=sr,n_fft=FRAME_LENGHT,hop_length=HOP_LENGTH,n_mfcc=64)
        delta_mfccs =  librosa.feature.delta(mfccs)
        librosa.display.specshow(mfccs, sr=sr, x_axis='time',y_axis="log")
        plt.colorbar(format='%+2.f')
    elif type == 8:
        mfccs = librosa.feature.mfcc(y=y, sr=sr,n_fft=FRAME_LENGHT,hop_length=HOP_LENGTH,n_mfcc=64)
        delta_mfccs =  librosa.feature.delta(mfccs,order=2) #delta delta 
        librosa.display.specshow(mfccs, sr=sr, x_axis='time',y_axis="log")
        plt.colorbar(format='%+2.f')
    elif type== 9:
        ae = amplitude_envelope_overlap(y)
        frames =  range(0,ae.size)
        t  = librosa.frames_to_time(frames,hop_length=HOP_LENGTH)
        plt.figure(1,figsize=(14,4),dpi=100)
        librosa.display.waveplot(y,alpha=0.5)
        plt.plot(t,ae,color="r")
        plt.legend(["Amplitude Envelope","Waveform"])

    else:
        plt.plot(y,alpha=0.5)
    plt.title(title)
    if(save):
        plt.savefig(f'{figpath}/Waveform.pdf', dpi=300)
    plt.show()
```

```python
def calculate_split_frequency_bin(split_frequency, sample_rate, num_frequency_bins):
    """Infer the frequency bin associated to a given split frequency."""
    import math
    frequency_range = sample_rate / 2
    frequency_delta_per_bin = frequency_range / num_frequency_bins
    split_frequency_bin = math.floor(split_frequency / frequency_delta_per_bin)
    return int(split_frequency_bin)


split_frequency_bin = calculate_split_frequency_bin(2000, 22050, 1025)
split_frequency_bin

```

```python
def band_energy_ratio(spectrogram, split_frequency, sample_rate):
    """Calculate band energy ratio with a given split frequency."""
    
    split_frequency_bin = calculate_split_frequency_bin(split_frequency, sample_rate, len(spectrogram[0]))
    band_energy_ratio = []
    
    # calculate power spectrogram
    power_spectrogram = np.abs(spectrogram) ** 2
    power_spectrogram = power_spectrogram.T
    
    # calculate BER value for each frame
    for frame in power_spectrogram:
        sum_power_low_frequencies = frame[:split_frequency_bin].sum()
        sum_power_high_frequencies = frame[split_frequency_bin:].sum()
        band_energy_ratio_current_frame = sum_power_low_frequencies / sum_power_high_frequencies
        band_energy_ratio.append(band_energy_ratio_current_frame)
    
    return np.array(band_energy_ratio)
```
